```python
#read and show image with cv2
image = cv2.imread("/content/1338290.png")

# Convert to grayscale (optional)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Feature extraction: Calculate color histogram
color_hist = cv2.calcHist([gray_image], [0], None, [256], [0, 256])  # Adjust parameters as needed

# Visualize histogram (optional)
plt.plot(color_hist)
plt.xlabel("Pixel Intensity")
plt.ylabel("Number of Pixels")
plt.title("Grayscale Color Histogram")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftshift
from scipy.io import loadmat
```

```python
bluewhale = loadmat('/content/bluewhale.mat')
bluewhale = bluewhale['bluewhale']
bluewhale.shape
```

```python
# read .mat file and plot
bluewhale = loadmat('/content/bluewhale.mat') #.csv, .wav,..
#here please look at structure of the original file to has the insight of the data
bluewhale = bluewhale['bluewhale']

# Visualize time-series data (heart rate vs time)
plt.plot(bluewhale)
plt.xlabel("Time (samples)")
plt.ylabel("Bluewhale sound")
plt.title("Bluewhale Time Series")
plt.show()
```

```python
# Fast Fourier Transform (FFT)

fft_data = fft(bluewhale)
fft_shifted = fftshift(abs(fft_data))  # Absolute value and shift for visualization
fs = 1/(bluewhale[2]-bluewhale[1]) #  sampling rate
# Sample frequencies based on data length
frequencies = np.arange(len(bluewhale)) / len(bluewhale) * (fs)
# Plot frequency spectrum
plt.plot(frequencies, fft_shifted)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title("Frequency Spectrum of Heart Rate Data")
plt.show()
```
